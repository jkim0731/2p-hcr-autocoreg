# Session 04 / R1 / Subgoal 01 (v2.1) — GFP+ threshold

**Goal.** Replace the hard `counts >= 5` GFP+ cutoff with a principled, **distribution-only** threshold. The legacy cutoff is both arbitrary and too low (54–68 % GFP+ on the four spot-data subjects), so R1's downstream estimators see far too many candidates.

**Constraints (from the user).**
1. The method must derive the threshold from the per-subject distribution alone. *No fixed percentile or fraction* — future subjects without a coreg table must inherit the same rule.
2. Among distribution-driven methods, prefer the one producing the **most stringent** GFP+ set (lowest `max gfp_frac` across subjects) while still covering ≥ 95 % of the subject's `coreg_table.csv` HCR IDs.
3. The `coreg_table` is used *only* to validate. No method consults it when picking the threshold.

**Scope.** Four spot-data subjects `788406, 790322, 767018, 782149`. Intensity-only subjects `755252, 767022` are out-of-scope — they carry no spot counts and will be handled in subgoal 02.

**v2.1 changes vs v2.** Density (`counts/volume`) strategies and joint `counts × density` strategies were added after the v2 rewrite. The joint Yen-on-counts × Yen-on-density method now wins, strictly tightening 767018 beyond what counts-only Yen achieves (9.5 % vs 10.9 %) with no regression on any other subject.

**Sources.** Driver `dev_code/04_r1_subgoal_01_gfp_threshold.py`, metrics `sessions/04_R1_coarse_align/subgoal_01_gfp_threshold_results.json`, figures `sessions/04_R1_coarse_align/subgoal_01_figures/`.

## 1. Why the baseline fails

`benchmark_data_loader._load_gfp()` applies `counts >= DEFAULT_GFP_MIN_SPOTS (=5)`, which sits at roughly the 25th percentile of spot-bearing cells. That puts **54–68 %** of *all* HCR cells into the GFP+ set — orders of magnitude above the expected sparse-label fraction and the main reason R1's 2-D GFP-density xcorr underperformed on XY translation (cf. `sessions/04_R1_coarse_align/log.md`).

The manual pipeline's heuristic (`min_counts = matched_cells.counts.min() * 1.2`) is circular at R1 (no matches exist yet) and cannot be reused.

## 2. Data characterisation

````python
from pathlib import Path
import json
import pandas as pd
from IPython.display import Image, Markdown, display

SESSION = Path('/root/capsule/code/sessions/04_R1_coarse_align')
FIG = SESSION / 'subgoal_01_figures'
with open(SESSION / 'subgoal_01_gfp_threshold_results.json') as f:
    res = json.load(f)
per_subj = res['per_subject']
SUBJECTS = ['788406','790322','767018','782149']
pd.set_option('display.float_format', lambda v: f'{v:.3f}')
````

````python
rows = []
for sid in SUBJECTS:
    d = per_subj[sid]
    row = {
        'subject': sid,
        'n_hcr_total': d['n_hcr_total'],
        'n_spot_bearing': d['n_spot_bearing_cells'],
        'spot_bearing_frac': d['n_spot_bearing_cells']/d['n_hcr_total'],
        'n_coreg_rows': d['n_coreg_rows'],
    }
    row.update({f'count_{k}': v for k, v in d['raw_counts_percentiles'].items()})
    if d.get('coreg_counts_percentiles'):
        row.update({f'coreg_{k}': v for k, v in d['coreg_counts_percentiles'].items()})
    rows.append(row)
pd.DataFrame(rows).set_index('subject')
````

**Observations.**
- Spot-bearing cells are **70–80 %** of all HCR cells — there is no clean noise-floor below `counts=1` because cells with 0 spots are not written to the CSV at all.
- The `coreg_table.csv` HCR IDs are the most trustworthy reference we have. Their count distribution has `p1` in the **tens** and `p5`/`p10` well into the hundreds, so any threshold up to ~100 is expected to preserve ≥95 % coverage.
- This means a distribution-based method can aim quite high (≥ 80 counts) before it starts dropping canonical pairs.
- The same structural story holds for *density* (`counts / volume`): no clean noise floor because 0-count cells are absent. Density therefore doesn't solve the threshold problem alone — but combined with counts it adds a second independent cut (see §6).

### 2.1 Coreg-cells vs all-HCR distribution (motivation for a high threshold)

Direct twin-axis histogram comparison on the same log axis: **left y-axis = all HCR cells** (grey), **right y-axis = canonical coreg-table cells only** (blue). Plotted for both `log10(counts)` and `log10(density)`, per subject.

The coreg-cell distribution is the gold-standard anchor set the threshold must preserve. Where it concentrates is the per-subject *safe tail* — a threshold placed below the coreg mode retains ≥ 95 % of coreg cells by construction, regardless of method. This is the visual intuition behind why Yen's ~80–160 count cutoffs work.

````python
import sys
sys.path.insert(0, '/root/capsule/code/dev_code')
import numpy as np
import matplotlib.pyplot as plt
from benchmark_data_loader import load_subject

def _raw_table(sid):
    s = load_subject(sid, gfp_threshold_method='counts_min', gfp_min_spots=1)
    df = s.hcr_gfp_df
    counts = df['counts'].values.astype(float)
    density = (df['density'].values.astype(float)
               if 'density' in df.columns else None)
    cg_ids = set(s.coreg_table['hcr_id'].astype(int).tolist())
    cg_mask = df['hcr_id'].astype(int).isin(cg_ids).values
    return counts, density, cg_mask

_cache = {sid: _raw_table(sid) for sid in SUBJECTS}

fig, axes = plt.subplots(2, 4, figsize=(17, 7.5))
for j, sid in enumerate(SUBJECTS):
    counts, density, cg = _cache[sid]
    # --- counts row
    ax = axes[0, j]
    ax2 = ax.twinx()
    c_log = np.log10(counts[counts > 0])
    cg_c_log = np.log10(counts[(counts > 0) & cg])
    bins_c = np.linspace(c_log.min(), c_log.max() + 0.05, 40)
    ax.hist(c_log, bins=bins_c, color='0.55', alpha=0.7, label=f'all (n={len(c_log)})')
    ax2.hist(cg_c_log, bins=bins_c, color='tab:blue', alpha=0.75, label=f'coreg (n={len(cg_c_log)})')
    ax.set_xlabel('log10(counts)')
    ax.set_ylabel('all HCR (left)', color='0.30')
    ax2.set_ylabel('coreg only (right)', color='tab:blue')
    ax.set_title(f'{sid} — counts')
    ax.legend(loc='upper left', fontsize=8)
    ax2.legend(loc='upper right', fontsize=8)
    # --- density row
    ax = axes[1, j]
    if density is None:
        ax.text(0.5, 0.5, 'no density column', ha='center', va='center', transform=ax.transAxes)
        ax.set_axis_off()
        continue
    ax2 = ax.twinx()
    pos = density > 0
    d_log = np.log10(density[pos])
    cg_d_log = np.log10(density[pos & cg])
    bins_d = np.linspace(d_log.min(), d_log.max() + 0.05, 40)
    ax.hist(d_log, bins=bins_d, color='0.55', alpha=0.7, label=f'all (n={len(d_log)})')
    ax2.hist(cg_d_log, bins=bins_d, color='tab:blue', alpha=0.75, label=f'coreg (n={len(cg_d_log)})')
    ax.set_xlabel('log10(density)')
    ax.set_ylabel('all HCR (left)', color='0.30')
    ax2.set_ylabel('coreg only (right)', color='tab:blue')
    ax.set_title(f'{sid} — density')
    ax.legend(loc='upper left', fontsize=8)
    ax2.legend(loc='upper right', fontsize=8)

fig.suptitle('Coreg cells (right axis, blue) vs all HCR ROIs (left axis, grey) — log-axis comparison', y=1.01)
fig.tight_layout()
plt.show()
````

## 3. Candidate strategies (all distribution-only)

Three families — counts-only, density-only, and joint counts×density.

| ID | Family | Description |
|---|---|---|
| baseline | reference | `counts >= 5` (not a candidate). |
| A | counts | 2-component GMM on `log(counts >= 1)`, posterior crossover at P=0.5. |
| A3 | counts | 3-component GMM on `log(counts)` with 0-count cells encoded as `log(0.5)` to admit a background mode. |
| B | counts | Otsu's between-class variance on `log(counts >= 1)`. |
| C | counts | Kneedle elbow of sorted-descending counts (below-chord maximum). |
| D | counts | Kittler–Illingworth min-error on `log(counts >= 1)`. |
| Triangle | counts | Zack's triangle method on `log(counts >= 1)`. |
| **Yen** | counts | Yen's maximum-entropy threshold on `log(counts >= 1)`. |
| ISODATA | counts | Iterative-mean (Ridler–Calvard) on `log(counts >= 1)`. |
| Yen/Otsu/Isodata/Triangle/GMM/GMM3bg *_log_density* | density | Same four distribution-driven methods applied to `log(density > 0)`. |
| **Joint_Yen × Yen** | joint | Keep cells passing both `counts ≥ t_c (Yen)` and `density ≥ t_d (Yen)`. |
| Joint_Otsu × Otsu | joint | Same idea with Otsu on both axes. |
| Joint_GMM × GMM | joint | Same idea with 2-component GMM on both axes. |

**Deliberately excluded:** fixed-fraction / fixed-percentile strategies ("top 15 %", "top 20 %", etc.). They require a hand-picked value that cannot be generalised to future subjects without a coreg table.

## 4. Per-strategy metrics

For each subject × strategy we record:
- `threshold_counts`: the integer counts cutoff (0 for density-only strategies).
- `threshold_density`: the density cutoff (None for counts-only strategies).
- `gfp_frac`: fraction of all HCR cells that pass (informational).
- `coreg_coverage`: fraction of coreg-table HCR IDs that pass — the primary quality metric. Target ≥ 0.95.

````python
def strategy_table(metric):
    rows = []
    names = list(per_subj[SUBJECTS[0]]['strategies'].keys())
    for name in names:
        row = {'strategy': name}
        for sid in SUBJECTS:
            row[sid] = per_subj[sid]['strategies'][name].get(metric)
        rows.append(row)
    return pd.DataFrame(rows).set_index('strategy')

display(Markdown('**threshold_counts** — per-subject integer counts cutoff'))
display(strategy_table('threshold_counts'))
display(Markdown('**threshold_density** — per-subject density cutoff (None for counts-only strategies)'))
display(strategy_table('threshold_density'))
display(Markdown('**gfp_frac** — |GFP+| / n_hcr_total'))
display(strategy_table('gfp_frac'))
display(Markdown(f'**coreg_coverage** — target ≥ {res["coverage_target"]:.2f}'))
display(strategy_table('coreg_coverage'))
````

## 6. Winner selection

Rule: of the distribution-driven strategies whose coreg coverage is ≥ 0.95 on **every** subject, pick the one with the lowest `max gfp_frac` across subjects — i.e., the most stringent GFP+ set. Tie-break by lower `mean gfp_frac`, then by higher `coreg_cov_min`. Baseline is excluded from the ranking (kept as a reference only).

````python
winners = res['winners_detail']
pd.DataFrame(winners).set_index('name')
````

### Verdict: **Joint_Yen_counts × Yen_density**

The joint Yen-on-counts × Yen-on-density rule produces the most stringent GFP+ set that still preserves ≥ 96 % canonical coreg coverage on every subject. Its `max gfp_frac` is the lowest of any passing strategy (**0.161** on 788406), with 767018 at **0.095** — noticeably tighter than pure Yen on counts (0.109).

**Why the joint cut helps on 767018.** A small population of 767018 cells has high raw counts but low density (low `counts/volume` — i.e. big, weakly-labelled cells). Pure counts-Yen lets them through; the density leg filters them out. On the other three subjects the density Yen cutoff is already cleared by virtually every cell that clears the counts cutoff, so the joint rule is identical to counts-only Yen there. The joint rule therefore dominates pure Yen: equal on 788406, 790322, 782149; strictly tighter on 767018.

**Per-subject comparison (joint vs pure Yen on counts):**

| subject | joint gfp_frac | pure-Yen gfp_frac | joint coreg covmin |
|---|---|---|---|
| 788406 | 0.161 | 0.161 | 0.985 |
| 790322 | 0.140 | 0.140 | 0.982 |
| 767018 | **0.095** | **0.109** | 0.996 |
| 782149 | 0.116 | 0.116 | 0.960 |

**Density-only strategies alone are too permissive.** Yen / Otsu / Isodata / GMM applied to `log(density > 0)` sit at 40–70 % gfp_frac because the density histogram (like counts) lacks a clean noise floor — 0-count cells don't appear. Density adds value only in combination with counts.

### Why GMM/Otsu/Triangle still under-cut on counts

The log-count distribution does not have two clean modes. The zero-count population is *absent* from the CSV (it's implicit), so methods that assume two populations **inside the spot-bearing data** find a split between weakly- and strongly-labelled cells — not between background and signal. That split sits at ~15–25 counts and is biologically too low. Triangle collapses to `counts = 2` because the histogram is monotonically decreasing. Yen works here because it maximises entropy across the log-count range, which on a heavy-tailed distribution naturally lands in the tail. The same structural issue explains why the density-only variants are too permissive — they see the same truncated histogram shape.

## 7. API wiring

`benchmark_data_loader` now defaults to `yen_log_joint`:

```python
load_subject(sid)                                          # yen_log_joint (default)
load_subject(sid, gfp_threshold_method='yen_log')          # counts-only Yen (diagnostic)
load_subject(sid, gfp_threshold_method='counts_min',       # legacy
                  gfp_min_spots=5)
load_subject(sid, gfp_threshold_method='fixed_frac',       # diagnostic only
                  gfp_target_frac=0.20)
```

The returned `SubjectData.gfp_min_spots` reflects the derived integer counts threshold (per subject); the density threshold is not persisted on the dataclass because the filter has already been applied. `gfp_threshold_method` records the name used. If a subject lacks a `density` column (e.g. aggregated spots without `metrics.pickle`), `yen_log_joint` falls back to counts-only Yen on that subject and logs no warning — the counts cut remains distribution-driven.
