# Iteration 1 — Intensity features at N22 surface vs v2.1 surface

**Goal.** Use N22's centroid-derived pia surface as per-subject ground truth and extract image features that separate the *real* pia from the autofluorescence-contaminated v2.1 surface on 755252.

**Inputs.** `data/iter01_features.csv` (2400 rows: 400 interior-grid columns × 6 subjects; feature set measured at both `n22` and `v21` surface z for each column); `data/iter01_profiles.npz` (a handful of full column profiles for plotting).

**Questions.**
1. Which features distinguish 755252's v21 position (AF top) from its N22 position (real pia)?
2. Do those features stay consistent across the other 5 subjects (where v21 ≈ N22)?

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

SESSION = Path('/root/capsule/code/sessions/03c_onset_features')
df = pd.read_csv(SESSION / 'data' / 'iter01_features.csv')
profiles = np.load(SESSION / 'data' / 'iter01_profiles.npz')
df['subject'] = df['subject'].astype(str)
print('rows:', len(df), 'subjects:', sorted(df.subject.unique()))
df.head(3)
```

## 1. Sample column profiles — see the AF signature

Plot three center-grid columns per subject with the N22 surface (green) and v2.1 surface (red) overlaid. On 755252 these should sit far apart; on the other five they should coincide.

```python
subjects = ['755252', '767018', '767022', '782149', '788406', '790322']
fig, axes = plt.subplots(2, 3, figsize=(15, 7), sharex=False, sharey=False)
for ax, sid in zip(axes.flat, subjects):
    prof = profiles[f'{sid}_center_prof']
    zax = profiles[f'{sid}_center_zaxis']
    z_n22, z_v21, xu, yu = profiles[f'{sid}_center_meta']
    ax.plot(zax, prof, 'k-', lw=0.8)
    ax.axvline(z_n22, color='green', lw=1.5, label=f'N22 z={z_n22:.0f}')
    ax.axvline(z_v21, color='red',   lw=1.5, label=f'v21 z={z_v21:.0f}')
    ax.set_title(f'{sid}  (µ at x={xu:.0f} y={yu:.0f})')
    ax.set_xlabel('z (µm)'); ax.set_ylabel('combined I(z)')
    ax.legend(fontsize=8, loc='upper right')
plt.tight_layout()
plt.savefig(SESSION / 'figures' / 'iter01_sample_profiles.png', dpi=120)
plt.show()
```

### Interpretation

On 755252 the v21 (red) line sits tens of µm **above** the N22 (green) line, landing on the top of an AF sheet. Shortly below v21 there is a **dark gap** before the real pia edge (green) marks the onset of bulk tissue signal. On all other subjects v21 and N22 essentially coincide.

The gap-below-AF pattern is the signal we want to detect image-only.

## 2. Summary table: feature medians at each surface

For every feature, per subject, compute median at `z_n22` and at `z_v21`. The delta (or ratio) is our signal: 755252 should stand out; others should be flat.

```python
feats = ['below_near', 'below_mid', 'below_deep',
         'dip_peak', 'dip_trough', 'dip_mag',
         'grad', 'above_near']

records = []
for f in feats:
    for sid in subjects:
        sub = df[df.subject == sid]
        n22_med = sub[f'n22_{f}'].median()
        v21_med = sub[f'v21_{f}'].median()
        records.append({
            'feature': f, 'subject': sid,
            'n22_med': n22_med, 'v21_med': v21_med,
            'delta': n22_med - v21_med,
        })
med_df = pd.DataFrame(records)
print('=== median N22  /  median v21 per subject ===\n')
print(med_df.pivot(index='feature', columns='subject', values='n22_med').round(4))
print('\n=== median v21 ===\n')
print(med_df.pivot(index='feature', columns='subject', values='v21_med').round(4))
print('\n=== delta (n22 - v21) ===\n')
print(med_df.pivot(index='feature', columns='subject', values='delta').round(4))
```

## 3. `below_near` — the best discriminator so far

`below_near = mean(I[z+10 µm : z+50 µm])` — mean combined intensity in a thin slab just below the candidate surface. Real pia should be bright here (tissue starts immediately). AF-top should be dim (the slab below the AF edge is the *gap* between AF and pia).

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
for ax, sid in zip(axes.flat, subjects):
    sub = df[df.subject == sid]
    ax.hist(sub['n22_below_near'], bins=40, alpha=0.6, color='green', label=f'at N22  med={sub.n22_below_near.median():.3f}')
    ax.hist(sub['v21_below_near'], bins=40, alpha=0.6, color='red',   label=f'at v21  med={sub.v21_below_near.median():.3f}')
    ax.set_title(sid); ax.set_xlabel('below_near (µ of I)'); ax.set_ylabel('count')
    ax.legend(fontsize=8)
plt.suptitle('Distribution of below_near at N22 vs v21 surface, per subject', y=1.02)
plt.tight_layout()
plt.savefig(SESSION / 'figures' / 'iter01_below_near_hist.png', dpi=120)
plt.show()
```

## 4. Scale-free variant: `below_near / below_deep`

Absolute `below_near` varies across subjects because tissue brightness does. A ratio to `below_deep` (bulk tissue at 150–250 µm) should be ≈ 1 at real pia (similar brightness) and ≈ 0 at AF-top (gap below, tissue further down).

```python
for loc in ['n22', 'v21']:
    df[f'{loc}_ratio_near_deep'] = df[f'{loc}_below_near'] / (df[f'{loc}_below_deep'].abs() + 1e-6)

fig, axes = plt.subplots(2, 3, figsize=(14, 8))
for ax, sid in zip(axes.flat, subjects):
    sub = df[df.subject == sid]
    ax.hist(sub['n22_ratio_near_deep'].clip(0, 3), bins=40, alpha=0.6, color='green',
            label=f'at N22 med={sub.n22_ratio_near_deep.median():.2f}')
    ax.hist(sub['v21_ratio_near_deep'].clip(0, 3), bins=40, alpha=0.6, color='red',
            label=f'at v21 med={sub.v21_ratio_near_deep.median():.2f}')
    ax.set_title(sid); ax.set_xlabel('below_near / below_deep'); ax.set_ylabel('count')
    ax.legend(fontsize=8)
plt.suptitle('Scale-free near/deep ratio', y=1.02)
plt.tight_layout()
plt.savefig(SESSION / 'figures' / 'iter01_ratio_near_deep_hist.png', dpi=120)
plt.show()
```

## 5. Discrimination summary

For each feature, compute a *separation score* = median-at-N22 / median-at-v21 (or if undefined, the absolute gap). A subject where v21 is correctly on pia should have score ≈ 1; one where v21 is on AF should deviate.

```python
rows = []
cands = ['below_near', 'below_mid', 'below_deep', 'ratio_near_deep']
for f in cands:
    for sid in subjects:
        sub = df[df.subject == sid]
        n22 = sub[f'n22_{f}'].median()
        v21 = sub[f'v21_{f}'].median()
        rows.append({'feature': f, 'subject': sid, 'n22': n22, 'v21': v21,
                     'abs_delta': abs(n22 - v21),
                     'ratio': (n22 / v21) if abs(v21) > 1e-6 else np.inf})
disc = pd.DataFrame(rows)
print(disc.pivot(index='feature', columns='subject', values='ratio').round(2))
print('\nabs_delta:')
print(disc.pivot(index='feature', columns='subject', values='abs_delta').round(3))
```

## Findings & next step

1. **`below_near` is the cleanest single feature.** At 755252's v21 the 40 µm slab below the surface is effectively zero (median 0.000) while at N22 it is 0.052 — a 50× deficit. The same test on the five "clean" subjects shows near-zero deltas.
2. **`below_near / below_deep`** is a scale-free version: the ratio is close to 1 at the real pia but collapses toward 0 at AF-tops.
3. **Dip-magnitude features were mostly zero**, because PBS and the AF gap both get clipped to 0 by the `load_hcr_combined` per-channel percentile normalisation (`clip(0, None)` after median subtraction). Iteration 2 should either work on raw per-channel intensities (no clipping) or craft a dip detector that tolerates the clip.

**Decision for iteration 2.** Use `below_near / below_deep` (or the raw `below_near` after switching to a non-clipping normalisation) as a per-column *pia confidence* score. Then rescore each candidate column top during the v2.1 fit so AF tops are suppressed. See `../log.md` iteration-1 entry.
