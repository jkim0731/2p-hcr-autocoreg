# Iteration 6 — OOT-aware scoring

**Goal.**  Plug an out-of-tissue (OOT) loophole in `score_surface_quality`.  The original implementation sampled a 20 × 20 interior grid without checking whether each column actually contained tissue.  Iter 5 tried to audit this with a `col.max() ≥ 5 % × vol.max()` rule and reported 52–94 % OOT on the first three HCR subjects, but that was a reference bug — see below.

**Change.**  `score_surface_quality` now takes `min_col_max_frac` (default 0.05) and an optional shared `global_max` / `global_max_percentile` kwargs.  Columns with `col.max() < min_col_max_frac × global_max` are dropped before taking medians; the scoring dict reports `n_oot` (number dropped).  `auto_select_surface` computes `global_max = np.percentile(ref_vol.max(axis=0), 99.5)` once and passes it to every candidate, so the mask is identical across the bank.

**Why `percentile(colmax, 99.5)` as the reference.**  The first attempt at the mask used `vol.max()` as the reference (what iter 5 reported).  On HCR `load_hcr_combined` output, `vol.max()` is dominated by a handful of saturated voxels — 32.8 on 755252 up to 253.6 on 788406 — 30–250 × a typical tissue column.  That inflated `5 % × vol.max()` above most real tissue columns (100 % OOT on 788406, 99.75 % on 790322 — the code raised `All-NaN slice encountered`).  Falling back to `np.percentile(vol, 99.5)` was the opposite failure: over the >1 B voxels of a level-4 volume the 99.5th percentile is a "typical bright voxel" (~0.3), not a "bright column" (~1.0), so every column including empty ones passed.  `np.percentile(colmax, 99.5)` splits the difference: robust to single-voxel saturation (each column contributes at most one bright-pixel's worth to `colmax`) and samples the correct scale (a bright tissue *column*).

**Revised understanding of iter 5's "52–94 % OOT".**  Under the correct reference the 15 %-padded 20 × 20 interior grid is mostly all-tissue:

| subject | in-tissue | OOT | global-max ref |
|---|---:|---:|---:|
| 755252 | 400 | 0 | 8.05 |
| 767018 | 400 | 0 | 4.79 |
| 767022 | 398 | 2 | 8.76 |
| 782149 | 391 | 9 | 6.00 |
| 788406 | 400 | 0 | 4.71 |
| 790322 | 400 | 0 | 5.57 |

So the "empty columns dragging Q/Qcon_n toward 0" story from iter 5 does not hold on this benchmark — the 15 %-padded interior grid avoids the side margins on all 6 subjects, and iter 3 / 4's low Q / Qcon_n values for non-pia candidates were legitimate pia rejection (not OOT contamination).

**Value of the fix.**  (i) Safety against saturated-voxel crashes under other choices of reference.  (ii) A well-defined mask for future data where the brain doesn't fill the FOV (e.g. narrower crops, partial volumes) — the machinery is in place.  (iii) Bottom-OOT doesn't need a separate mask: a surface pushed below the tissue produces `Qcon_n ≈ -1` (below is zero-padded, above is tissue) and is rejected by the existing contrast gate.

**Validation.**  Re-ran the 7-candidate bank on all 6 HCR subjects.  Selections unchanged vs iter 4 on 6/6.

## Per-subject candidate tables (iter 6 scoring)

Same 7-candidate bank as iter 4; `Q` and `Qcon_n` are now computed on in-tissue grid columns only.  `n_oot` is the number of grid columns dropped.  `above` / `onset` are centroid-truth validation (not used by the selection rule).

```python
import numpy as np
df = pd.read_csv(SESSION / 'data' / 'iter06_autoselect.csv')
df['subject'] = df.subject.astype(str)
for sid in subjects:
    sub = df[df.subject == sid].sort_values('Q', ascending=False).copy()
    sub['mark'] = np.where(sub.selected, '★',
                  np.where(sub['pass'], '✓', '✗'))
    print(f'\n=== {sid} ===')
    print(sub[['mark', 'name', 'Q', 'Qabove', 'Qcon_n', 'n', 'n_oot', 'above', 'onset']]
          .to_string(index=False, float_format=lambda x: f'{x:7.4f}'))
```

## Bottom line

- Iter 5's headline "52–94 % OOT" was a reference-inflation artefact.  `vol.max()` is dominated by saturated voxels on HCR `load_hcr_combined` output (32–253 × typical), so `5 % × vol.max()` was inflated above most real tissue columns.  Under the robust reference `np.percentile(colmax, 99.5)` the mask drops 0–9 of 400 grid columns per subject — an order of magnitude less than iter 5 claimed.
- Auto-select still correctly agrees with iter 4 on 6/6 subjects.  Onset = 2.5 µm and above-fraction ≤ 1.13 % hold.
- The fix is kept as a safety measure: with the naive reference `vol.max()` the threshold is inflated so high on 788406 / 790322 that *every* grid column gets filtered out as OOT, and the medians raise `All-NaN slice encountered`.  With the robust reference the mask is a mild no-op on this benchmark but gives the scoring a principled tissue gate for future data (narrower FOVs, partial crops, etc.).
- Bottom-OOT (padded zeros below the tissue) does not need a dedicated check: a surface pushed below the tissue reads `Qcon_n ≈ -1` and is rejected by the existing contrast gate.
- Iter 5's qualitative check (surfaces overlay correctly on every subject, v2.1 orange on 755252 sits ~40 µm above the pia) still stands — that part of the audit was independent of the OOT reference.
