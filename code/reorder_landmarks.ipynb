# Purpose:
- Reorder landmarks to find the landmarks in the viewer
- For additional manual correction after the first iteration


```python
import pandas as pd
import os
```

```python
mouse_id = 767018
iter=3
```

```python
landmark_file = f'/root/capsule/code/{mouse_id}_coreg/{mouse_id}_landmarks_matched_ext_iter{iter}.csv'
landmarks = pd.read_csv(landmark_file, header=None)
columns = ['ids', 'active', 'czstack_x', 'czstack_y', 'czstack_z', 'hcr_x', 'hcr_y', 'hcr_z']
landmarks.columns = columns

# convert czstack ids to 04d format
def _format_id(x):
    if x.startswith('cz'):
        cz_id = int(x.split('-')[0].split('cz')[-1])
        return f'cz{cz_id:04d}-{x.split("-")[-1]}'
    return x
landmarks['ids'] = landmarks['ids'].apply(_format_id)

landmarks.sort_values(by='ids', inplace=True)
landmarks.head(25)
```

```python
save_fn = f'/root/capsule/code/{mouse_id}_coreg/{mouse_id}_landmarks_matched_ext_iter{iter}_reordered.csv'
assert not os.path.exists(save_fn), f'File already exists: {save_fn}'
landmarks.to_csv(save_fn, index=False, header=False)
```
